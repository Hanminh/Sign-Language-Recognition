```python
import numpy as np
import os
import matplotlib.pyplot as plt
root = "Data_debug"
loss_history = []
for path in os.listdir(root):
    loss_path = os.path.join(root, path)
    loss = np.load(loss_path)
    for i in range(len(loss)):
        loss_history.append(loss[i])
plt.plot(loss_history)
```
